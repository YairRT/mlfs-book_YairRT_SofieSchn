```python
import hopsworks

project = hopsworks.login()
mr = project.get_model_registry()

model_names = [
    'air_quality_xgboost_model_1',
    'air_quality_xgboost_model_2',
    'air_quality_xgboost_model_3',
    'air_quality_xgboost_model_4',
    'air_quality_xgboost_model_5'
]

for name in model_names:
    # Get all model versions for this name
    models = mr.get_models(name)
    for model in models:
        print(f"Deleting model: {model.name}, version: {model.version}")
        model.delete()
```
